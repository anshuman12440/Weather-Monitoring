# STEC TO VTEC conversion


```python
import math
R_e = 6371
H_i =350
data_gps_new['VTEC'] = data_gps_new['TEC_TOW'].astype(float)*math.cos(math.asin(R_e/(R_e + H_i))*(math.cos(el_cutoff)))
```

# IPP Calculation


```python
data_gps_new
```

```python
def ipp_cal(azimuth):
    latitude = math.radians(75.8577)
    longitude = math.radians(22.7196)

    psi_pp  = math.pi/2  - el_cutoff - math.asin((R_e*math.cos(el_cutoff))/(R_e +H_i))
    phi_pp =  math.asin(math.sin(latitude)*math.cos(psi_pp) + math.cos(latitude)*math.sin(psi_pp)*math.cos(azimuth))
    lamda_pp  = longitude + math.asin((math.sin(psi_pp)*math.sin(azimuth)/math.cos(phi_pp)))
    return lamda_pp
    
                                                
```

```python
data_gps_new['Azimuth'] =data_gps_new['Azimuth'].astype(float)
```

```python

data_gps_new['ipp_val'] = data_gps_new.apply(lambda x :ipp_cal(x['Azimuth']),axis =1)

```
