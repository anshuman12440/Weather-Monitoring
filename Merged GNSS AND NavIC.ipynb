# Calendar to Julian

```python
def cal2julian(yr,mn,dy):
    if mn < 1 | mn > 12:
        print('Invalid input month')
        return



    if dy < 1:
        print('Invalid input day')
        return
        
        
    if (mn == 2 and dy > 29) or (((mn== 3) or mn==5 or mn==9 or mn ==11)and dy > 30) or (dy > 31):
        print('Invalid input day');

        return


    if mn > 2:
        y = yr
        m = mn

    else:
        y = yr - 1
        m = mn + 12



    date1=4.5+31*(10+12*1582)

    date2=15.5+31*(10+12*1582) 

    date=dy+31*(mn+12*yr)

    if date <= date1:
        b = -2

    elif date >= date2:
        b = np.fix(y/400) - np.fix(y/100)

    else:
        print('Dates between October 5 & 15, 1582 do not exist')
        return



    if y > 0:
        jd = np.fix(365.25*y) + np.fix(30.6001*(m+1)) + b + 1720996.5 + dy
        return jd

    else:
        
        jd= np.fix(365.25*y-0.75) + np.fix(30.6001*(m+1)) + b + 1720996.5 + dy
        return jd


```

# NavIC to julian

```python
jdNavIC = cal2julian(1999,8,22)

def navic2jd(week,sow,rollover=0):
    nweek = week + 1024*rollover
    jd = jdNavIC + nweek*7 + (sow/3600)/24
    return jd

navic2jd(1011,259200)
    
```

# julain to cal

```python
import math
from datetime import datetime
def jd_to_date(jd):
   
   
    jd = float(jd) + 0.5
    
    F, I = math.modf(jd)
    I = int(I)
    
    A = math.trunc((I - 1867216.25)/36524.25)
    
    if I > 2299160:
        B = I + 1 + A - math.trunc(A / 4.)
    else:
        B = I
        
    C = B + 1524
    
    D = math.trunc((C - 122.1) / 365.25)
    
    E = math.trunc(365.25 * D)
    
    G = math.trunc((C - E) / 30.6001)
    
    day = C - E + F - math.trunc(30.6001 * G)
    
    if G < 13.5:
        month = G - 1
    else:
        month = G - 13
        
    if month > 2.5:
        year = D - 4716
    else:
        year = D - 4715
        Day= int(day)
    Hour = (day-int(day))*24
    Minute = (Hour-int(Hour))*60
    sec = (Minute-int(Minute))*60
    t = year,month,Day,Hour,Minute,sec
    
    return "%d-%d-%d %d:%d:%d" %t

    
```
