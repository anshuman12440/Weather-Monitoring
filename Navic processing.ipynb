```python
import numpy as np
import glob
import tkinter as tk
from tkinter import filedialog 
import csv
```

```python
root = tk.Tk()
root.withdraw()

file_path = filedialog.askdirectory()
```

```python
path = file_path+'/*/*.csv'
files = glob.glob(path)
sep_path=[]
for line in files:
    temp = line.split('\\')
    temp[1] = float(temp[1])
    sep_path.append(temp)
sep_path = sorted(sep_path,key=lambda x: (x[2],x[1]))
files =[]
for i in range(0,len(sep_path)):
    temp = sep_path[i]
    files.append(temp[0]+'/'+str(int(temp[1]))+'/'+temp[2])
```

```python
dt=[]
list_3d = []
for i in range(0,int(len(files)/8)):
    filename = sep_path[i*8][2]
    data=[]
    for j in range(0,8):
        file = files[i*8 + j]
        count = 0
        with open(file,'r',errors='ignore') as reader:
            for line in reader.readlines():
                if(count!=0):
                    data.append(line.split(',')[0:-1])
                else:
                    count=count+1
                    header = line.split(',')[0:-1]
    list_3d.append(data)
    dt.append([filename,header])
```

```python
for i in range(0,len(list_3d)):
    
    with open(sep_path[0][0]+'/'+dt[i][0], 'w') as f:
      
    # using csv.writer method from CSV package
        write = csv.writer(f)
      
        write.writerow(dt[i][1])
        write.writerows(list_3d[i])
```

```python
import numpy as np
def cal2julian(yr,mn,dy):
    if mn < 1 | mn > 12:
        print('Invalid input month')
        return



    if dy < 1:
        print('Invalid input day')
        return
        
        
    if (mn == 2 and dy > 29) or (((mn== 3) or mn==5 or mn==9 or mn ==11)and dy > 30) or (dy > 31):
        print('Invalid input day');

        return


    if mn > 2:
        y = yr
        m = mn

    else:
        y = yr - 1
        m = mn + 12



    date1=4.5+31*(10+12*1582)

    date2=15.5+31*(10+12*1582) 

    date=dy+31*(mn+12*yr)

    if date <= date1:
        b = -2

    elif date >= date2:
        b = np.fix(y/400) - np.fix(y/100)

    else:
        print('Dates between October 5 & 15, 1582 do not exist')
        return



    if y > 0:
        jd = np.fix(365.25*y) + np.fix(30.6001*(m+1)) + b + 1720996.5 + dy
        return jd

    else:
        
        jd= np.fix(365.25*y-0.75) + np.fix(30.6001*(m+1)) + b + 1720996.5 + dy
        return jd


```

```python
import math
def jd_to_date(jd):
   
   
    jd = jd + 0.5
    
    F, I = math.modf(jd)
    I = int(I)
    
    A = math.trunc((I - 1867216.25)/36524.25)
    
    if I > 2299160:
        B = I + 1 + A - math.trunc(A / 4.)
    else:
        B = I
        
    C = B + 1524
    
    D = math.trunc((C - 122.1) / 365.25)
    
    E = math.trunc(365.25 * D)
    
    G = math.trunc((C - E) / 30.6001)
    
    day = C - E + F - math.trunc(30.6001 * G)
    
    if G < 13.5:
        month = G - 1
    else:
        month = G - 13
        
    if month > 2.5:
        year = D - 4716
    else:
        year = D - 4715
    Day= int(day)
    Hour = (day-int(day))*24
    Minute = (Hour-int(Hour))*60
    sec = (Minute-int(Minute))*60
        
    return year, month, Day,Hour,Minute,sec
    
```
